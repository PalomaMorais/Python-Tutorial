# Plotando gráficos científicos com o Python

A biblioteca básica para plotagem no Python é a `matplotlib`, mas ela pode ser um pouco complicada para iniciantes, por isso não cobrimos ela durante a aula. 

Neste documento, vamos destrinchar algumas funcionalidades e mostrar como fazer gráficos mais profissionais com o Python.

1. Primeiramente, vamos importar a biblioteca `matplotlib`. Ela é dividade em vários módulos, mas a seção que realmente faz as plotagens em alto nível é a `pyplot`. Importaresmos o `pandas` também para lidar com os dados. 

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Vamos ler os dados

```python
df = pd.read_csv('car_data.csv',sep=';')
df.head()
```

Lembrem-se que podemos plotar gráficos diretamente com o `pandas`. Vamos plotar uma dispersão.

*obs: `plt.show()` é o comando para mostrar o gráfico. Ele não é necessário em todas as situações, mas é melhor previnir*

```python
df.plot(x='hp', y='km/l-urb', kind='scatter')
plt.show()
```

Vamos aumentar o tamanho da figura

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))
plt.show()
```

Os nomes dos eixos estão com os nomes das colunas do banco de dados, mas podemos mudar esses nomes para algo mais explicativo através da biblioteca `matplotlib`. Vamos usar `xlabel` e `ylabel` com uma fonte de tamanho 16 e tipo `Times New Roman` 

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#mostra o gráfico
plt.show()
```

Podemos adicionar linhas de grade no gráfico com a função `plt.grid()`. Podemos colocar grade ao longo do eixo x (`x`), do eixo y (`y`) e dos dois `both`. Além disso, diminuímos a opacidade das linhas através do parâmetro `alpha`.

*obs: alpha é um parâmetro que varia de 0 a 1 e representa a opacidade variando desde uma linha sólida (alpha=1) até uma linha completamente transparente (alpha=0)*

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#grade
plt.grid(axis='both', alpha=.3)

#mostra o gráfico
plt.show()
```

Vamos modificar os limites do gráfico, forçando que x varie de 0 a 300 e y de 4 a 20.

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#grade
plt.grid(axis='both', alpha=.3)

#limites
plt.xlim(0,300)
plt.ylim(4,20)

#mostra o gráfico
plt.show()
```

Vamos diminuir o número de linhas de grade no eixo y. Para isso vamos usar a função `range()` que retorna os valores de um número a outro com saltos fixos. a sintaxe é range(valor inicial, valor final, salto). vamos usar um salto de 4 unidades

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#grade
plt.grid(axis='both', alpha=.3)

#limites
plt.xlim(0,300)
plt.ylim(4,20)

#linhas de grade
plt.yticks(range(4,21,4))

#mostra o gráfico
plt.show()
```

Podemos também mudar a cor de fundo do gráfico

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#grade
plt.grid(axis='both', alpha=.3)

#limites
plt.xlim(0,300)
plt.ylim(4,20)

#linhas de grade
plt.yticks(range(4,21,4))

#mostra o gráfico
plt.show()
```

Vamos remover a envoltória do gráfico

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#grade
plt.grid(axis='both', alpha=.5)

#limites
#vamos aumentar o limite superior um pouco para tornar a linha de x=300 visível 
plt.xlim(0,301)

#vamos baixar o limite inferior um pouco para tornar a linha de y=4 visível 
plt.ylim(3.9,20)

#linhas de grade
plt.yticks(range(4,21,4))

#remove envoltória
plt.box(on=None)

#mostra o gráfico
plt.show()
```

## Utilizando estilos

Para usar estilos diferentes vamos usar o módulo `style` para alterar todo o estilo do gráfico

```python
import matplotlib.style as style
```

Existem vários estilos para o `matplotlib`. Para maiores detalhes ver https://matplotlib.org/gallery/style_sheets/style_sheets_reference.html

Vamos usar o estilo `ggplot` baseado no livro **Grammar of Graphics**

```python
style.use('ggplot')
```

Agora podemos plotar nosso gráfico como anteriormente

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6))

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#grade
plt.grid(axis='both', alpha=1)

#limites
#vamos aumentar o limite superior um pouco para tornar a linha de x=300 visível 
plt.xlim(0,301)

#vamos baixar o limite inferior um pouco para tornar a linha de y=4 visível 
plt.ylim(3.9,20)

#linhas de grade
plt.yticks(range(4,21,4))

#mostra o gráfico
plt.show()
```

Parece bom, mas podemos fazer mais alguns ajustes. Vamos aumentar o tamanho dos pontos, dar transparência a eles e mudar a cor.

Para uma lista de cores, ver:
https://matplotlib.org/examples/color/named_colors.html

```python
df.plot(x='hp', y='km/l-urb', kind='scatter', figsize=(10,6), s=40, alpha=.6, color='k')

#nomes dos eixos
plt.xlabel('Potência do Motor (hp)', fontsize=16, fontname='Times New Roman')
plt.ylabel('Rendimento (km/l de combustível)', fontsize=16, fontname='Times New Roman')

#grade
plt.grid(axis='both', alpha=1)

#limites
#vamos aumentar o limite superior um pouco para tornar a linha de x=300 visível 
plt.xlim(0,301)

#vamos baixar o limite inferior um pouco para tornar a linha de y=4 visível 
plt.ylim(4,20)

#linhas de grade
plt.yticks(range(4,21,4))

#mostra o gráfico
plt.show()
```
